```python
import nmrglue as ng
import numpy as np
import pandas as pd

# ...existing code including nmr_df...

def export_2d_flat(data_path, export_path):
    """
    Exports 2D NMR data as a flat table with columns for F1 ppm, F2 ppm, F1 Hz, F2 Hz, and intensity.
    
    Parameters:
    data_path (str): Path to the NMR data.
    export_path (str): File path for the exported CSV.
    """
    dic, data = ng.bruker.read_pdata(data_path)
    udic = ng.bruker.guess_udic(dic, data)
    ndim = udic['ndim']
    if ndim != 2:
        raise ValueError("This export function is only for 2D NMR data.")
    
    # Get the frequency information for both dimensions.
    uc_f1 = ng.fileiobase.uc_from_udic(udic, dim=0)
    uc_f2 = ng.fileiobase.uc_from_udic(udic, dim=1)
    ppm_f1 = uc_f1.ppm_scale()
    ppm_f2 = uc_f2.ppm_scale()
    hz_f1 = uc_f1.hz_scale()
    hz_f2 = uc_f2.hz_scale()
    
    # Create meshgrids for each frequency axis.
    # Note: Using 'ij' indexing so that the shapes match the original data dimensions.
    F1_ppm, F2_ppm = np.meshgrid(ppm_f1, ppm_f2, indexing='ij')
    F1_hz, F2_hz = np.meshgrid(hz_f1, hz_f2, indexing='ij')
    
    # Flatten the arrays to create a table.
    df_flat = pd.DataFrame({
        'F1 (ppm)': F1_ppm.flatten(),
        'F2 (ppm)': F2_ppm.flatten(),
        'F1 (Hz)': F1_hz.flatten(),
        'F2 (Hz)': F2_hz.flatten(),
        'intensity': data.flatten()
    })
    
    # Export the flat table to CSV.
    df_flat.to_csv(export_path, index=False)
```

```python
export_2d_flat('data/2D/36/pdata/1', '2d_data.csv')
```

```python
# Read csv and make countour plot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('2d_data.csv')
data_grid = df.pivot(index='F1 (ppm)', columns='F2 (ppm)', values='intensity')

F1 = data_grid.index.values
F2 = data_grid.columns.values
X, Y = np.meshgrid(F2, F1)  # F2 as x-axis and F1 as y-axis
Z = data_grid.values
```

```python
# reset plt.ylim to nothing
```

```python
contour_levels = 5e5 * 1.2 ** np.arange(15)

# Create the .contour plot
plt.figure(figsize=(10, 8))
plt.contour(X, Y, Z, levels=contour_levels, cmap='viridis')
plt.xlim(180, -10)
plt.ylim(16, -5)
```
